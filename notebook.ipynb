```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Generate a random 2D dataset

np.random.seed(0)
X = np.random.rand(100, 2)
y = X[:, 0] + X[:, 1] > 1

# 2. plot the dataset
plt.scatter(X[y == 0, 0], X[y == 0, 1], c='red', marker='o', label='class 0')
```
